# MODEL TRAINING

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
```

```python
def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np) - window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(target, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.1)
X1_train.shape, X1_test.shape, y1_train.shape, y1_test.shape
```

```python
# X1_train = X1_train.reshape((X1_train.shape[0], X1_train.shape[1], X1_train.shape[3]))
# X1_test = X1_test.reshape((X1_test.shape[0], X1_test.shape[1], X1_test.shape[3]))
```

## MODEL 1


```python
model1 = Sequential()
model1.add(InputLayer((5, 1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python
cp1 = ModelCheckpoint('models/model1/m1.keras', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X1_train, y1_train, epochs=70, validation_data=(X1_test, y1_test))
```

```python
# from tensorflow.keras.models import load_model
# model1 = load_model('models/model1/m1.keras')
```

```python
train_predictions = model1.predict(X1_train)
train_predictions_flat = train_predictions.flatten()
y1_train_flat = y1_train.flatten()

# Create a DataFrame
train_results = pd.DataFrame(data={'Train Predictions': train_predictions_flat, 'Actuals': y1_train_flat})

# Display the DataFrame
print(train_results.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train Predictions'][50:100])
plt.plot(train_results['Actuals'][50:100])
```

```python
# Test results
val_predictions = model1.predict(X1_test).flatten()
val_results = pd.DataFrame(data={'Val Predictions':val_predictions, 'Actuals':y1_test})
val_results
```

```python
from sklearn.metrics import mean_squared_error as mse

def plot_predictions1(model, X, y, start=0, end=100):
  predictions = model.predict(X).flatten()
  df = pd.DataFrame(data={'Predictions': predictions, 'Actuals':y})
  plt.plot(df['Predictions'][start:end])
  plt.plot(df['Actuals'][start:end])
  return df, mse(predictions, y)
```

## MODEL 2

```python
model2 = Sequential()
model2.add(InputLayer((WINDOW_SIZE, 1)))
model2.add(Conv1D(64, kernel_size=2, activation='relu'))
model2.add(Flatten())
model2.add(Dense(8, 'relu'))
model2.add(Dense(1, 'linear'))
model2.summary()
```

```python
cp2 = ModelCheckpoint('models/model2/m2.keras', save_best_only=True)
model2.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model2.fit(X1_train, y1_train, validation_data=(X1_test, y1_test), epochs=50, callbacks=[cp2])
```

```python
plot_predictions1(model2, X1_test, y1_test)
```

```python
new_df = final_data[['Year', 'R-(Sunspot No.)', 'f10.7_index', 'TEC']]
new_df
```

```python
def df_to_X_y2(df, window_size = 6):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [r for r in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size][0]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
NEW_WINDOW_SIZE = 6
X2, y2 = df_to_X_y2(new_df, NEW_WINDOW_SIZE)
X2.shape, y2.shape
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.1)
X2_train.shape, X2_test.shape, y2_train.shape, y2_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the training data
X2_train_normalized = scaler.fit_transform(X2_train.reshape(-1, X2_train.shape[-1])).reshape(X2_train.shape)

# Transform the test data
X2_test_normalized = scaler.transform(X2_test.reshape(-1, X2_test.shape[-1])).reshape(X2_test.shape)

y2_train_normalized = scaler.fit_transform(y2_train.reshape(-1, 1)).reshape(y2_train.shape)
y2_test_normalized = scaler.transform(y2_test.reshape(-1, 1)).reshape(y2_test.shape)
```

```python
X2_train_normalized.shape, X2_test_normalized.shape, y2_train_normalized.shape, y2_test_normalized.shape
```

```python
# without normalising
model4 = Sequential()
model4.add(InputLayer((NEW_WINDOW_SIZE, 4)))
model4.add(LSTM(64))
model4.add(Dense(8, 'relu'))
model4.add(Dense(1, 'linear'))

model4.summary()
```

```python
cp4 = ModelCheckpoint('models/model4/m4.keras', save_best_only=True)
model4.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.001), metrics=[RootMeanSquaredError()])
```

```python
model4.fit(X2_train_normalized, y2_train_normalized, validation_data=(X2_test_normalized, y2_test_normalized), epochs=10, callbacks=[cp4])
```
